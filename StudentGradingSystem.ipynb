```python
students = []
#Student Records
def add_student(name, marks):
    record = {
        "name": name,
        "marks": marks,
        "total": sum(marks),
        "percentage": sum(marks) / len(marks)
    }

    record["grade"] = calculate_grade(record["percentage"])
    students.append(record)


# Grading
def calculate_grade(percent):
    if percent >= 80:
        return "A"
    elif percent >= 70:
        return "B"
    elif percent >= 60:
        return "C"
    elif percent >= 50:
        return "D"
    else:
        return "F"


#Searching algorithm
def search_student(name):
    for s in students:
        if s["name"].lower() == name.lower():
            return s
    return None


# Sorting algorithm
def sort_by_percentage():
    return sorted(students, key=lambda x: x["percentage"], reverse=True)


# Statistics
def class_stats():
    if not students:
        return None

    avg = sum(s["percentage"] for s in students) / len(students)

    highest = max(students, key=lambda x: x["percentage"])
    lowest = min(students, key=lambda x: x["percentage"])

    return avg, highest, lowest


# Display
def show_all():
    for s in students:
        print(s["name"], s["percentage"], s["grade"])



```

```python
add_student("Prist", [78, 80, 90])
add_student("Hari", [60, 65, 70])
add_student("Shreya", [40, 50, 45])

print("All Students")
show_all()

print("\nSearch Hari")
print(search_student("Hari"))

print("\nSorted")
print(sort_by_percentage())

print("\nStatistics")
print(class_stats())
```
