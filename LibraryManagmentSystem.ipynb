```python
class Book:
    def __init__(self, book_id, title, author):
        self.__book_id = book_id
        self.__title = title
        self.__author = author
        self.__is_available = True

    def get_id(self):
        return self.__book_id

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

    def is_available(self):
        return self.__is_available

    def borrow(self):
        if self.__is_available:
            self.__is_available = False
            return True
        return False

    def return_book(self):
        self.__is_available = True

    def display(self):
        status = "Available" if self.__is_available else "Borrowed"
        print(self.__book_id, self.__title, self.__author, status)
```

```python
class User:
    def __init__(self, user_id, name):
        self._user_id = user_id
        self._name = name

    def get_name(self):
        return self._name

    def get_id(self):
        return self._user_id

    def get_role(self):
        return "User"
```

```python
class Student(User):
    def get_role(self):
        return "Student"


class Librarian(User):
    def get_role(self):
        return "Librarian"
```

```python
class Library:
    def __init__(self):
        self.books = []
        self.users = []

    # -------- Book Operations --------

    def add_book(self, book):
        self.books.append(book)

    def find_book(self, book_id):
        for book in self.books:
            if book.get_id() == book_id:
                return book
        return None

    def show_books(self):
        for book in self.books:
            book.display()

    # -------- User Operations --------

    def add_user(self, user):
        self.users.append(user)

    # -------- Borrow / Return --------

    def borrow_book(self, book_id):
        book = self.find_book(book_id)

        if book is None:
            print("Book not found")
            return

        if book.borrow():
            print("Book borrowed successfully")
        else:
            print("Book is already borrowed")

    def return_book(self, book_id):
        book = self.find_book(book_id)

        if book is None:
            print("Book not found")
            return

        book.return_book()
        print("Book returned successfully")
```

```python
library = Library()

# Add Books
library.add_book(Book(1, "Python Basics by", "Prist"))
library.add_book(Book(2, "Data Structures by ", "Hari"))
library.add_book(Book(3, "OOP Design by ", "Shreya"))

# Add Users
library.add_user(Student(101, "Gyanu"))
library.add_user(Librarian(201, "Maya"))

print("All Books")
library.show_books()

print("\nBorrow Book 1")
library.borrow_book(1)

print("\nAfter Borrow")
library.show_books()

print("\nReturn Book 1")
library.return_book(1)

print("\nFinal Status")
library.show_books()
```
